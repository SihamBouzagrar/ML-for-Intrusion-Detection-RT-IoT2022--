<h2 id="understanding_data">Understanding the Data</h2>

### `RT_IOT2022`:
We have downloaded a , **`RT_IOT2022`**, which is derived from a real-time IoT infrastructure and contains malicious network traffic generated by various IoT devices and attack scenarios. [Dataset source](https://archive.ics.uci.edu/dataset/942/rt-iot2022)

- **id.orig_p** The network port used by the origin source
- **id.resp_p** The network port used by the responding device
- **proto**The protocol utilized in the connection
- **service**The type of service used in the connection
- **flow_duration**How long the connection lasted between devices
- **fwd_pkts_tot** Total packets forwarded in the session
- **bwd_pkts_tot**Total packets backward (returned) in the session
- **fwd_data_pkts_tot**Total data packets forwarded in the session (excluding control packets)
- **bwd_data_pkts_tot** Total data packets backward in the session (excluding control packets)
- **fwd_pkts_per_sec** Rate of packets forwarded per second
- **bwd_pkts_per_sec** Rate of packets backward per second
- **flow_pkts_per_sec** Combined rate of packets per second
- **down_up_ratio**Ratio of downstream to upstream traffic
- **fwd_header_size_tot**Total size of the header forwarded
- **fwd_header_size_min**Smallest header size sent
- **fwd_header_size_max**Largest header size sent
- **bwd_header_size_tot**Total size of the header returned
- **bwd_header_size_min**Smallest header size returned
- **fwd_header_size_max**Largest header size returned
-**flow_FIN_flag_count**Count of FIN flags signaling the end of data transmission in a session
- **flow_SYN_flag_count** : Count of SYN flags used to initiate and establish sessions  
- **flow_RST_flag_count** : Count of RST flags used to abruptly terminate sessions  

- **fwd_PSH_flag_count** : Count of PSH flags in forward packets indicating the push of buffered data to the receiving application  
- **bwd_PSH_flag_count** : Count of PSH flags in backward packets indicating the push of buffered data to the sending application  

- **flow_ACK_flag_count** : Count of ACK flags used to acknowledge the receipt of packets  

- **fwd_URG_flag_count** : Count of URG flags in forwarded packets indicating data should be processed urgently  
- **bwd_URG_flag_count** : Count of URG flags in backward packets indicating data should be processed urgently  

- **flow_CWR_flag_count** : Count of CWR (Congestion Window Reduced) flags used by the sender to signal congestion control  
- **flow_ECE_flag_count** : Count of ECE (ECN Echo) flags indicating network congestion without dropping packets  

- **fwd_pkts_payload.min** : Minimum payload size in forwarded packets  
- **fwd_pkts_payload.max** : Maximum payload size in forwarded packets  
- **fwd_pkts_payload.tot** : Total payload size in forwarded packets  
- **fwd_pkts_payload.avg** : Average payload size in forwarded packets  
- **fwd_pkts_payload.std** : Standard deviation of payload sizes in forwarded packets  

- **bwd_pkts_payload.min** : Minimum payload size in backward packets  
- **bwd_pkts_payload.max** : Maximum payload size in backward packets  
- **bwd_pkts_payload.tot** : Total payload size in backward packets  
- **bwd_pkts_payload.avg** : Average payload size in backward packets  
- **bwd_pkts_payload.std** : Standard deviation of payload sizes in backward packets  

- **flow_pkts_payload.min** : Minimum payload size in the flow  
- **flow_pkts_payload.max** : Maximum payload size in the flow  
- **flow_pkts_payload.tot** : Total payload size in the flow  
- **flow_pkts_payload.avg** : Average payload size in the flow  
- **flow_pkts_payload.std** : Standard deviation of payload sizes in the flow  

- **fwd_iat.min** : Minimum inter-arrival time of forwarded packets  
- **fwd_iat.max** : Maximum inter-arrival time of forwarded packets  

- **fwd_iat.tot** : Total inter-arrival time of forwarded packets  
- **fwd_iat.avg** : Average inter-arrival time of forwarded packets  
- **fwd_iat.std** : Standard deviation of inter-arrival times of forwarded packets  

- **bwd_iat.min** : Minimum inter-arrival time of backward packets  
- **bwd_iat.max** : Maximum inter-arrival time of backward packets  
- **bwd_iat.tot** : Total inter-arrival time of backward packets  
- **bwd_iat.avg** : Average inter-arrival time of backward packets  
- **bwd_iat.std** : Standard deviation of inter-arrival times of backward packets  

- **flow_iat.max** : Maximum inter-arrival time in the flow  
- **flow_iat.tot** : Total inter-arrival time in the flow  
- **flow_iat.std** : Standard deviation of inter-arrival times in the flow  

- **payload_bytes_per_second** : Rate of payload transmission in bytes per second  

- **fwd_subflow_pkts** : Forward subflow packet count  
- **bwd_subflow_pkts** : Backward subflow packet count  
- **bwd_subflow_bytes** : Backward subflow byte count  

- **fwd_bulk_bytes** : Forward bulk byte count  
- **bwd_bulk_bytes** : Backward bulk byte count  
- **fwd_bulk_packets** : Forward bulk packet count  
- **bwd_bulk_packets** : Backward bulk packet count  
- **fwd_bulk_rate** : Rate of bulk data transmission in the forward direction  
- **bwd_bulk_rate** : Rate of bulk data transmission in the backward direction  

- **active.min** : Minimum time the flow was active before going idle  
- **active.max** : Maximum time the flow was active before going idle  
- **active.tot** : Total time the flow was active before going idle  
- **active.avg** : Average time the flow was active before going idle  
- **active.std** : Standard deviation of active times before the flow went idle  

- **idle.min** : Minimum time the flow was idle  
- **idle.max** : Maximum time the flow was idle  
- **idle.tot** : Total time the flow was idle  
- **idle.avg** : Average time the flow was idle  
- **idle.std** : Standard deviation of idle times  

- **fwd_init_window_size** : Initial window size in forwarded TCP connections, indicative of congestion handling  
- **bwd_init_window_size** : Initial window size in backward TCP connections, indicative of congestion handling  
- **fwd_last_window_size** : Last window size observed in forwarded TCP connections, can indicate changes in network conditions  

- **Attack_type** : Specifies the type of network traffic, whether normal or related to specific types of network attacks

To get an overview of the features with a strong correlation the following listing shows all feature correlations with an absolute value > 0.8.

```python
def get_abs_correlations(RT_IOT2022, threshold=0.8):
    corr = RT_IOT2022.corr(numeric_only=True)
    cols = corr.columns
    strong_corr = set()

    for i in range(len(cols)):
        for j in range(i):
            strong_corr.add((cols[i], cols[j]))

    c = (
        corr.abs()
        .unstack()
        .drop(labels=strong_corr)
        .sort_values(ascending=False)
    )

    return c[c > threshold]

with pd.option_context('display.max_rows', None):
    print(get_abs_correlations(RT_IOT2022, 0.8))
```

```python
# Convert the categorical columns to nuMeric
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
RT_IOT2022['Attack_type_encoded'] = labelencoder.fit_transform(RT_IOT2022['Attack_type'])
attack_mapping = dict(zip(labelencoder.classes_, labelencoder.transform(labelencoder.classes_)))
print("Mapping Type d'attaque → Encodage :", attack_mapping)
```

```python
#Encodage par OneHotEncoder de deux attributs categorielles 'proto' et 'service'
from sklearn.preprocessing import OneHotEncoder
features = ['proto', 'service']
ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoded = ohe.fit_transform(RT_IOT2022[features])
encoded_df = pd.DataFrame(
    encoded,
    columns=ohe.get_feature_names_out(features),
    index=RT_IOT2022.index
)
RT_IOT2022 = RT_IOT2022.drop(columns=features)
RT_IOT2022 = pd.concat([RT_IOT2022, encoded_df], axis=1)
RT_IOT2022
```

### Justification de l’utilisation de la PCA

Compte tenu :
- des **fortes corrélations entre plusieurs variables explicatives**,
- de la **redondance d’information** observée,
- et du **nombre élevé de features**,

l’application de l’**Analyse en Composantes Principales (PCA)** est **bénéfique** dans ce cas.  
Elle permet de :
- réduire la dimensionnalité du jeu de données,
- limiter les effets de la multicolinéarité,
- tout en conservant l’essentiel de la variance informative.


Train/ Test split

The split is stratified using the attack category in order to guarantee that all attacks are represented in the training and test set based on their occurences in the dataset.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,stratify=y, random_state=42)
```

```python
print("size of dataset : ", RT_IOT2022.shape)
print(f"size of training dataset : {X_train.shape}")
print("size of test dataset : ", X_test.shape)
```

```python
y_test.value_counts()
```

```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE

# Downsample la classe majoritaire
under = RandomUnderSampler(
    sampling_strategy='majority',
    random_state=42
)

X_train_under, y_train_under = under.fit_resample(X_train, y_train)

# Oversample les classes minoritaires
smote = SMOTE(random_state=42)


X_train_bal, y_train_bal = smote.fit_resample(
    X_train_under, y_train_under
)
```

```python
def detect_all_outliers_tukey(X_train_scaled_df):
    outliers_dict = {}

    # Sélection uniquement des colonnes numériques
    numeric_cols = X_train_scaled_df.select_dtypes(include='number').columns

    for col in numeric_cols:
        x = pd.to_numeric(X_train_scaled_df[col], errors='coerce').dropna()

        q1 = x.quantile(0.25)
        q3 = x.quantile(0.75)
        iqr = q3 - q1

        floor = q1 - 1.5 * iqr
        ceiling = q3 + 1.5 * iqr

        outliers =X_train_scaled_df[(X_train_scaled_df[col] < floor) | (X_train_scaled_df[col] > ceiling)][col]

        if len(outliers) > 0:
            outliers_dict[col] = pd.DataFrame({
                'indice': outliers.index,
                'value': outliers.values
            })

    return outliers_dict
```

```python
outliers = detect_all_outliers_tukey(X_train_scaled_df)
```

```python
sorted_outliers = sorted(
    outliers.items(),
    key=lambda item: len(item[1]),
    reverse=True
)

# Affichage
for col, data in sorted_outliers:
    print(f"{col} : {len(data)} outliers")
```

VI. Mise en place du pipeline et préparation au déploiement


```python
import pickle
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from xgboost import XGBClassifier
from sklearn.decomposition import PCA

# Colonnes
num_cols = X.select_dtypes(include=["int64","float64"]).columns
cat_cols = X.select_dtypes(include=["object"]).columns

preprocessor = ColumnTransformer([
    ("num", StandardScaler(), num_cols),
    ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols)
])

pca = PCA(n_components=31, random_state=42)

model = XGBClassifier(
    learning_rate=0.1,
    max_depth=5,
    n_estimators=200,
    eval_metric="mlogloss",
    random_state=42
)

# ===== SMOTE uniquement pour train =====
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X_train, y_train)

# ===== pipeline SANS smote =====
pipeline = Pipeline([
    ("preprocess", preprocessor),
    ("pca", pca),
    ("classifier", model)
])

pipeline.fit(X_res, y_res)

# Label encoder
le = LabelEncoder()
le.fit(y_res)
attack_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print("Mapping Type d'attaque → Encodage :", attack_mapping)

with open("pipeline1.pkl","wb") as f:
    pickle.dump(pipeline,f)

with open("final_model1.pkl","wb") as f:
    pickle.dump(le,f)
```
